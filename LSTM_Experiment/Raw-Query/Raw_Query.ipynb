```python
# Query Data
from prometheus_api_client import PrometheusConnect
import pandas as pd
from datetime import datetime
from zoneinfo import ZoneInfo

# Connect to Prometheus
prom = PrometheusConnect(url = "http://10.35.29.108:31102/", disable_ssl = True)

# Define Time Range
start_time_utc_unaware = datetime(2026, 2, 6, 0, 0, 0)
end_time_utc_unaware   = datetime(2026, 2, 12, 0, 0, 0)

print(f"Fetching data from {start_time_utc_unaware} to {end_time_utc_unaware}...")

print(int(start_time_utc_unaware.timestamp()))

print(int(end_time_utc_unaware.timestamp()))

query_cpu_max = 'kube_node_status_capacity{resource="cpu", node!~".*master.*|.*control-plane.*"}'
data_cpu_max = prom.custom_query_range(query = query_cpu_max, start_time = start_time_utc_unaware, end_time = end_time_utc_unaware, step ='1m')

query_cpu_request = 'kube_pod_container_resource_requests{resource="cpu", node!~".*master.*|.*control-plane.*"}'
data_cpu_request = prom.custom_query_range(query = query_cpu_request, start_time = start_time_utc_unaware, end_time = end_time_utc_unaware, step = '1m')

query_pending_pod = 'kube_pod_status_phase{phase="Pending"}'
data_pending_pod = prom.custom_query_range(query = query_pending_pod, start_time = start_time_utc_unaware, end_time = end_time_utc_unaware, step = '1m')

query_pod_info = 'kube_pod_info{node!~".*master.*|.*control-plane.*"}'
data_pod_info = prom.custom_query_range(query = query_pod_info, start_time = start_time_utc_unaware, end_time = end_time_utc_unaware, step = '1m')

data_cpu_max_pd = pd.DataFrame(data_cpu_max)
data_cpu_request_pd = pd.DataFrame(data_cpu_request)
data_pending_pod_pd = pd.DataFrame(data_pending_pod)
data_pod_info_pd = pd.DataFrame(data_pod_info)

data_cpu_max_pd.to_csv(f'Raw_CPU_Max.csv', index = False)
data_cpu_request_pd.to_csv('Raw_CPU_Request.csv', index = False)
data_pending_pod_pd.to_csv('Raw_Pending_Pod.csv', index = False)
data_pod_info_pd.to_csv('Raw_Pod_Info.csv', index = False)

# sum(max by (cluster, namespace, pod, container)(cluster:namespace:pod_cpu:active:kube_pod_container_resource_requests{cluster="", node=~"aj-aung-k8s-worker1"})) by (pod)
```
