```python
# Query Data
from prometheus_api_client import PrometheusConnect, MetricRangeDataFrame
from datetime import datetime
import pandas as pd
import zoneinfo

# Connect to Prometheus
prom = PrometheusConnect(url = "http://10.35.29.108:31102", disable_ssl = True)

# Define Time Range
thai_tz = zoneinfo.ZoneInfo("Asia/Bangkok")

# Create datetime objects localized to Thai time
start_time_thai = datetime(2026, 2, 28, 14, 0, 0, tzinfo = thai_tz)
end_time_thai = datetime(2026, 3, 3, 14, 0, 0, tzinfo = thai_tz)

print(f"Fetching data from {start_time_thai} to {end_time_thai}...")
print(int(start_time_thai.timestamp()))
print(int(end_time_thai.timestamp()))
```

```python
print(prom.all_metrics())
```

```python
cpu_req_query       = 'sum(max by (cluster, namespace, pod, container)(cluster:namespace:pod_cpu:active:kube_pod_container_resource_requests{cluster="", node=~"(aj-aung-k8s-worker1|aj-aung-k8s-worker2)"})) by (pod)'
cpu_cap_query       = 'sum(kube_node_status_capacity{cluster="", job="kube-state-metrics", node=~"(aj-aung-k8s-worker1|aj-aung-k8s-worker2)", resource="cpu"})'
mem_req_query       = 'sum(max by (cluster, namespace, pod, container)(node_namespace_pod_container:container_memory_working_set_bytes{cluster="", node=~"(aj-aung-k8s-worker1|aj-aung-k8s-worker2)", container!=""})) by (pod)'
mem_cap_query       = 'sum(kube_node_status_capacity{cluster="", job="kube-state-metrics", node=~"(aj-aung-k8s-worker1|aj-aung-k8s-worker2)", resource="memory"})'
number_of_pod_query = 'sum(kube_pod_owner{job="kube-state-metrics", cluster="", node=~"(aj-aung-k8s-worker1|aj-aung-k8s-worker2)"}) by (namespace)'
number_of_pending_pod_query = 'sum(kube_pod_status_phase{job="kube-state-metrics", cluster="", node=~"(aj-aung-k8s-worker1|aj-aung-k8s-worker2)", phase="Pending"}) by (namespace)'
```

```python
cpu_req_pd = MetricRangeDataFrame(prom.custom_query_range(query = cpu_req_query, start_time = start_time_thai, end_time = end_time_thai, step = "1m"))
cpu_cap_pd = MetricRangeDataFrame(prom.custom_query_range(query = cpu_cap_query, start_time = start_time_thai, end_time = end_time_thai, step = "1m"))
mem_req_pd = MetricRangeDataFrame(prom.custom_query_range(query = mem_req_query, start_time = start_time_thai, end_time = end_time_thai, step = "1m"))
mem_cap_pd = MetricRangeDataFrame(prom.custom_query_range(query = mem_cap_query, start_time = start_time_thai, end_time = end_time_thai, step = "1m"))
number_of_pod_pd = MetricRangeDataFrame(prom.custom_query_range(query = number_of_pod_query, start_time = start_time_thai, end_time = end_time_thai, step = "1m"))
number_of_pending_pod_query_pd = MetricRangeDataFrame(prom.custom_query_range(query = number_of_pending_pod_query, start_time = start_time_thai, end_time = end_time_thai, step = "1m"))
```
