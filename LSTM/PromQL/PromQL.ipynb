```python
# Query Data
from prometheus_api_client import PrometheusConnect, MetricRangeDataFrame
from datetime import datetime
import pandas as pd
import zoneinfo

# Connect to Prometheus
prom = PrometheusConnect(url = "http://10.35.29.108:31102", disable_ssl = True)

# Define Time Range
thai_tz = zoneinfo.ZoneInfo("Asia/Bangkok")

# Create datetime objects localized to Thai time
start_time_thai = datetime(2026, 2, 8, 0, 0, 0, tzinfo = thai_tz)
end_time_thai = datetime(2026, 2, 13, 12, 0, 0, tzinfo = thai_tz)

print(f"Fetching data from {start_time_thai} to {end_time_thai}...")
print(int(start_time_thai.timestamp()))
print(int(end_time_thai.timestamp()))
```

```python
print(prom.all_metrics())
```

```python
query = 'sum(max by (cluster, namespace, pod, container)(cluster:namespace:pod_cpu:active:kube_pod_container_resource_requests{cluster="", node=~"(aj-aung-k8s-worker1)"})) by (pod)'

data = prom.custom_query_range(query = query, start_time = start_time_thai, end_time = end_time_thai, step = "1m")

data_pd = MetricRangeDataFrame(data)
```
